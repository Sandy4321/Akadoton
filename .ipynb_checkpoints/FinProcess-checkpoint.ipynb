```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
import os 
from tqdm import tqdm
import gc
```

```python
%%time
folder = './data_akado/'

abonent = pd.read_csv(folder+'Абоненты.txt', sep='	', encoding='cp1251', decimal=',')
# adress_problem = pd.read_csv(folder+'адреса по проблемам.txt', sep='	', encoding='cp1251', decimal=',')
# services_activation = pd.read_csv(folder+'Активизация и Аннуляция услуг.txt', sep='	', encoding='cp1251', decimal=',')
active_users = pd.read_csv(folder+'Активные абоненты.txt', sep='	', encoding='cp1251', decimal=',')
# action = pd.read_csv(folder+'Акции.txt', sep='	', encoding='cp1251', decimal=',')
# smet_atributes = pd.read_csv(folder+'Атрибуты смет.txt', sep='	', encoding='cp1251', decimal=',')
# problems_groups = pd.read_csv(folder+'Группы услуг по проблемам.txt', sep='	', encoding='cp1251', decimal=',')
# history_services = pd.read_csv(folder+'История услуг.txt', sep='	', encoding='cp1251', decimal=',')
# smet_transactions = pd.read_csv(folder+'начисления по сметам.txt', sep='	', encoding='cp1251', decimal=',')
# services_transactions = pd.read_csv(folder+'Начисления по услугам.txt', sep='	', encoding='cp1251', decimal=',')
user_msgs = pd.read_csv(folder+'обращения.txt', sep='	', encoding='cp1251', decimal=',')
user_payments = pd.read_csv(folder+'Платежи.txt', sep='	', encoding='cp1251', decimal=',')
# user_problems = pd.read_csv(folder+'проблемы.txt', sep='	', encoding='cp1251', decimal=',')
# smet = pd.read_csv(folder+'сметы.txt', sep='	', encoding='cp1251', decimal=',')
# actions_info = pd.read_csv(folder+'Справочник Акций.txt', sep='	', encoding='cp1251', decimal=',')
# smet_atributions_info = pd.read_csv(folder+'Справочник Атрибутов смет.txt', sep='	', encoding='cp1251', decimal=',')
# reasons_info = pd.read_csv(folder+'Справочник причин изменений.txt', sep='	', encoding='cp1251', decimal=',')
# servies_info = pd.read_csv(folder+'Справочник Услуг.txt', sep='	', encoding='cp1251', decimal=',')
# services = pd.read_csv(folder+'Услуги.txt', sep='	', encoding='cp1251', decimal=',')
```

```python
abonent = abonent[['Код абонента', 'Район', 'АО']]
```

```python
# make only for enthernet task
active_users = active_users[active_users['Группа услуг']=='Интернет подключения'].reset_index(drop=True)

# determine start of 'month axis'
start_month = pd.to_datetime('2014-12-01 00:00:00')
```

## user payments 

```python
user_payments['Дата платежа'] = user_payments['Дата платежа'].apply(lambda x: x.split(' ')[0])
user_payments['Дата платежа'] = pd.to_datetime(user_payments['Дата платежа'], format='%d.%m.%Y')
user_payments['idx_month'] = (user_payments['Дата платежа'].dt.year - start_month.year)*12 + user_payments['Дата платежа'].dt.month - start_month.month
```

## making target 

```python
# target 
active_users = active_users.sort_values(by='Месяц активности').reset_index(drop=True)
active_users['Месяц активности'] = pd.to_datetime(active_users['Месяц активности'], format='%Y-%m')
active_users['idx_month'] = (active_users['Месяц активности'].dt.year - start_month.year)*12 + active_users['Месяц активности'].dt.month - start_month.month
```

## Working with user msgs 

```python
user_msgs['Дата обращения'] = user_msgs['Дата обращения'].apply(lambda x: x.split(' ')[0])
user_msgs['Дата обращения'] = pd.to_datetime(user_msgs['Дата обращения'], format='%d.%m.%Y')
user_msgs['idx_month'] = (user_msgs['Дата обращения'].dt.year - start_month.year)*12 + user_msgs['Дата обращения'].dt.month - start_month.month
```

# make train 

```python
def string_combiner(x):
    x = x.values
    x = [str(t) for t in x]
    return ' '.join(x)
```

```python
np.min(active_users['idx_month']), np.max(active_users['idx_month'])
```

```python
Data = pd.DataFrame({})

for DT in tqdm(np.arange(5, 33, 1)):
    next_month = DT + 3
    
    # make target 
    train_part = active_users[(active_users['idx_month']>=DT) & (active_users['idx_month']<=(DT+2))]
    target_part = active_users[active_users['idx_month']==next_month]
    num_monthes = train_part[['Код абонента', 'idx_month']].groupby('Код абонента').agg('count').reset_index()
    num_monthes['idx_month'] = num_monthes['idx_month']/3
    new_cols = list(num_monthes.columns)
    new_cols[-1] = 'frac_monthes'
    num_monthes.columns = new_cols
    train_part = pd.merge(train_part, num_monthes, how='left', on='Код абонента')
    train_part = train_part[train_part['frac_monthes']==1].reset_index(drop=True)
    next_month_users = set(target_part['Код абонента'])
    train_part['target'] = train_part['Код абонента'].apply(lambda x: 1*(x not in next_month_users))
    
    train_part['target_month'] = target_part['Месяц активности'][-1:].values[0] # for seasonality feature 
    
    # number of monthes in use 
    train_part_temp = active_users[active_users['idx_month']<=(DT+2)]
    temp = train_part_temp[['Код абонента', 'idx_month']].groupby(['Код абонента']).agg('count').reset_index()
    new_cols = list(temp.columns)
    new_cols[-1] = 'monthes_in_use'
    temp.columns = new_cols
    train_part = pd.merge(train_part, temp, how='left', on='Код абонента')
    
    # msgs from user 
    user_msgs_temp = user_msgs[(user_msgs['idx_month']>=DT) & (user_msgs['idx_month']<=(DT+2))]
    user_msgs_temp = user_msgs_temp[['Код абонента', 'Код обращения из Справочника']].groupby('Код абонента').agg(string_combiner).reset_index()
    train_part = pd.merge(train_part, user_msgs_temp, how='left', on='Код абонента')
    train_part['Код обращения из Справочника'] = train_part['Код обращения из Справочника'].fillna('')
    
    # number of msgs 
    train_part['nmbr_msgs'] = train_part['Код обращения из Справочника'].apply(lambda x: len(x.split(' ')))
    train_part['nmbr_uniq_msgs'] = train_part['Код обращения из Справочника'].apply(lambda x: len(set(x.split(' '))))
    train_part['frac_uniq_msgs'] = train_part['nmbr_uniq_msgs'] / (train_part['nmbr_uniq_msgs']+1)
    
    # payment for last 3 monthes 
    user_payments_temp = user_payments[(user_payments['idx_month']>=DT) & (user_payments['idx_month']<=(DT+2))]
    user_payments_temp = user_payments_temp[['Код абонента', 'Сумма платежа']].groupby(['Код абонента']).agg('sum').reset_index()
    train_part = pd.merge(train_part, user_payments_temp, how='left', on='Код абонента')
    
    # concat in one DF 
    train_part = train_part.drop(['frac_monthes', 'Группа услуг'], axis=1)
    Data = pd.concat([Data, train_part], axis=0)
    
    del train_part
    gc.collect()

Data = Data.reset_index(drop=True)
```

## model 

1 - Churn; 0 - Not churn

|   | Churn  | Not churn  | 
|---|---|---|
| Positive prediction  |  True positive |  False positive | 
| Negative prediction  | False negative  | True negative  | 

___

Удержание пользователя - 200 уе
Потеря пользователя (потеря прибыли за 3 мес) - 1000

|   | Churn  | Not churn  | 
|---|---|---|
| Positive prediction  |  -200 |  -200 | 
| Negative prediction  | -1000  | 0  | 

$Utility = churn.rate*((TPR*TP_{ev}) + (1-TPR)*FN_{ev}) + (1-churn.rate)*(FPR*FP_{ev} + (1-FPR)*FP_{ev})$

```python
# custom metric 
from sklearn.metrics import roc_curve, auc

churn_fraction = 0.05

uder_pols = -200
churn_ev = -1000
tp_ev = uder_pols
fp_ev = uder_pols
fn_ev = churn_ev
tn_ev = 0

def EV_calc(tpr, fpr, thresholds, churn_fraction):
    EV_model = []
    for i in range(len(thresholds)):
        ev = (tpr[i]*tp_ev + (1-tpr[i])*fn_ev)*churn_fraction + ((1-fpr[i])*tn_ev + fpr[i]*fp_ev)*(1-churn_fraction)
        EV_model.append(ev)
    return EV_model

EV_noModel = EV_calc([0], [0], [0.5], churn_fraction) # do nothing 
EV_allKeep = EV_calc([churn_fraction], [1-churn_fraction], [0.5], churn_fraction) # all keep
EV_idealModel = EV_calc([1], [0], [0.5], churn_fraction)

const_utility = np.max([EV_noModel[0], EV_allKeep[0]])
print([EV_noModel[0], EV_allKeep[0]])
print('const_utility', const_utility, 'ideal', EV_idealModel )


def utility(preds, dtrain):
    labels = dtrain.get_label()
    preds = preds
    
    fpr, tpr, thresholds = roc_curve(labels, preds)
    roc_auc0 = auc(fpr, tpr)
    EV = EV_calc(tpr, fpr, thresholds, churn_fraction)
    EV = np.max(EV)
    return 'utility', EV-const_utility, True

```

```python
import lightgbm as lgb
from sklearn.model_selection import GroupKFold

gkf = GroupKFold(n_splits=5)
```

```python
train_ds = lgb.Dataset(X_train, Train['target'], free_raw_data=False)
```

```python
params = {
    'objective':'binary', 
    'metric': 'AUC',
    'learning_rate': 0.1024, 
    'random_state':4242442,
    'verbose_eval': 0,
    'scale_pos_weight':25
}
h = lgb.cv(params, train_ds, verbose_eval=10, num_boost_round=999999, feval=utility, early_stopping_rounds=20,
           folds=gkf.split(Train['target'], Train['target'], Train['idx_month']))
```

```python
model = lgb.train(params, train_ds, num_boost_round=1700)
```
