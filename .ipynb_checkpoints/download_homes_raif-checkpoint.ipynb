```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
```

```python
train = pd.read_csv('train_set.csv')
```

```python
train = train[['home_add_lat','home_add_lon']]
```

```python
from math import sin, cos, sqrt, atan2, radians

moscow_center_lat, moscow_center_lon = 55.751244, 37.618423

# approximate radius of earth in km
R = 6373.0

lat1 = moscow_center_lat
lon1 = moscow_center_lon
lat2 = train['home_add_lat']
lon2 = train['home_add_lat']

dlon = abs(lon2 - lon1)
dlat = abs(lat2 - lat1)

a = dlon * 6 + dlat*11

train['dist_moscow'] = a
```

```python
print(train.shape)
#train = train[train.dist_moscow<110]
train = train.drop('dist_moscow', axis=1)
train = train.drop_duplicates().reset_index(drop=True)
print(train.shape)
```

```python
import folium
m = folium.Map(location=list(train.iloc[20]))

points = train[-50:].values

points_tuple = []
for i in points:
    points_tuple.append(tuple(list(i)))

for i in points:
    folium.Marker(list(i), icon=folium.Icon(icon='cloud')).add_to(m)
```
